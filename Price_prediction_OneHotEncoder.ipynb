# **Price Prediction with Dummy Variable**

```python
import pandas as pd
```

```python
df=pd.read_csv(r"/content/homeprices (1).csv")
```

```python
df
```

```python
dummy = pd.get_dummies(df.town)
dummy
```

```python
merged_df = pd.concat([df,dummy],axis='columns')
merged_df
#created a dummy table for towns using pd.get_dummy()
#merged_df is new merged dataframe having both original dataframe and dummy town df
# now we will remove town column from merged data frame
```

```python
final = merged_df.drop(['town'],axis='columns')
final
```

```python
# now we can remove any of three to reduce the computation as we can take either robinville as 1 or west windsor as 1 and if both are 0 then it indicates that it is monroe township
# remove either west windsor or robinsville or monroe township from final

final = final.drop(['west windsor'],axis='columns')
final
```

```python
X=final.drop(['price'],axis='columns')
X
```

```python
y=final.price
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```

```python
model.predict([[3400,0,0]])     # price of 3400 sqft in west windsor
```

```python
model.predict([[2800,0,1]])     # price of 2800 sqft in robinsville
```

```python
model.predict([[2500,1,0]])     #price of 2500 sqft in monroe township
```
