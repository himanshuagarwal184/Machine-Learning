```python
import numpy as np
```

```python
def gradient_descent(x,y):
  m_curr=b_curr=0       #m is slope and b is constant i.e mx+b
  iterations = 10000
  learning_rate=.08
  n=len(x)
  for i in range(iterations):
    y_pred = m_curr * x + b_curr
    cost = (1/n)*sum([val**2 for val in (y-y_pred)])
    md = -(2/n)*sum(x*(y-y_pred))     # differentiate equation  y= 1/n Î£ i=1 to n (yi-(mxi+b))^2  This is equation of mean squared error
    bd = -(2/n)*sum(y-y_pred)
    m_curr = m_curr-md*learning_rate
    b_curr = b_curr-bd*learning_rate
    print(m_curr,"\t",b_curr,"\t",cost,"\t",i)
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
