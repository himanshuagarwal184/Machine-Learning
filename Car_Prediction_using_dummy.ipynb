```python
import pandas as pd
```

```python
df = pd.read_csv(r"/content/carprices.csv")
df
```

```python
dummy_df = pd.get_dummies(df['Car Model'])
dummy_df
```

```python
merged_df = pd.concat([df,dummy_df],axis='columns')
merged_df
```

```python
final = merged_df.drop(['Car Model'],axis='columns')
final
```

```python
y=final['Sell Price($)']
y
```

```python
final = final.drop(['Sell Price($)','Mercedez Benz C class'],axis='columns')
final
```

```python
X=final
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(X,y)
```
